```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Erstellung eines leeren DataFrames
data = []

# Reale Namen von Berliner Klubs
club_names = ['Berghain', 'Watergate', 'KitKatClub', 'Tresor', 'Sisyphos', 'Ipse', 'About Blank', 'Renate', 'Salon zur Wilden Renate', 'Kater Blau']

# Dummy-Daten für Klubbesuche generieren
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 12, 31)

for _ in range(2000):
    klubname = random.choice(club_names)
    datum = start_date + timedelta(days=random.randint(0, 365))
    besuchername = random.choice(['Anna', 'Max', 'Sophie', 'Paul', 'Lena'])
    besucherzahl = random.randint(50, 300)
    eintrittspreis = round(random.uniform(5.0, 20.0), 2)
    bewertung = random.randint(1, 5)
    data.append([klubname, datum, besuchername, besucherzahl, eintrittspreis, bewertung])

# Erstellen eines DataFrames aus der Liste
df = pd.DataFrame(data, columns=['Klubname', 'Datum', 'Besuchername', 'Besucherzahl', 'Eintrittspreis', 'Bewertung'])

# Daten in eine CSV-Datei schreiben
df.to_csv('dummy_klubbesuche.csv', index=False)
```
