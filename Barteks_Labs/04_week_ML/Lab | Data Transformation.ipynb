```python
import pandas as pd
df = pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
df.head()
```

```python
# creating categorical_df
categorical_df = df.select_dtypes(include=['object', 'category'])
```

```python
categorical_df
```

```python
unique_values_region = df['region'].unique()
unique_values_response = df['response'].unique()
unique_values_coverage = df['coverage'].unique()
unique_values_education = df['education'].unique()
unique_values_month = df['month'].unique()
unique_values_employment_status = df['employment_status'].unique()
```

```python
print("Unique values in region:", unique_values_region)
print("Unique values in response:", unique_values_response)
print("Unique values in coverage:", unique_values_coverage)
print("Unique values in education:", unique_values_education)
print("Unique values in month:", unique_values_month)
print("Unique values in employment_status:", unique_values_employment_status)
```

```python
# Hot/Label Encoding of the categorical variable "response"
```

```python
# endocing nominal variables with get_dummies 
nominals_df_coding = pd.get_dummies(categorical_df[["response", "region", "employment_status", 
                                             "effective_to_date", "gender", "location_code", "marital_status"
                                             , "policy_type", "policy", "renew_offer_type"
                                             , "sales_channel", "vehicle_class", "vehicle_size"]], dtype = int)
nominals_df_coding
```

```python
# endocing ordinal variables with get_dummies
ordinal_df_coding_coverage = categorical_df["coverage"].map({"basic":0, "extended": 1,"extended": 2})
ordinal_df_coding_education = categorical_df["education"].map({"college":1, "bachelor": 2,"high school or below": 0,"master": 3,"doctor": 4 })
ordinal_df_coding_month = categorical_df["month"].map({"feb":0, "jan": 1})
```

```python
ordinal_df_coding = pd.DataFrame({
    'coverage': ordinal_df_coding_coverage,
    'education': ordinal_df_coding_education,
    'month': ordinal_df_coding_month
})

ordinal_df_coding 
```

```python
categorical_df
```

```python
# CONCATINATING nominal and ordinal 

# Concatenate along columns (axis=1)
categorical_combined_df = pd.concat([ordinal_df_coding , nominals_df_coding], axis=1)


# final variable for encoded catergorical values 
categorical_combined_df
```

```python
# APPLIYING MIN MAX SCALER 

scaler_claim_amount = MinMaxScaler()
scaler_claim_amount.fit(numerical_df[["months_since_last_claim"]]) # defintes the range, finds min and max and creates a frame for the scale
scaler_claim_amount.transform(numerical_df[["months_since_last_claim"]]) # applies formula to each value, amends data to scale 

scaler_claim_amount
```

```python
scaled_values = scaler_claim_amount.fit_transform(numerical_df[["months_since_last_claim"]])
scaled_df = pd.DataFrame(scaled_values, columns=["months_since_last_claim"])

# Display the histogram
scaled_df["months_since_last_claim"].hist()
```
