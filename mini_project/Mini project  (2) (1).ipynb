```python
import pandas as pd
df = pd.read_csv("Bundesliga_Results.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Filter rows where the month is September and Bayern Munich is the home or away team
filtered_data = df[(df['Date'].dt.month == 9) & ((df['HomeTeam'] == 'Bayern Munich') | (df['AwayTeam'] == 'Bayern Munich'))]

filtered_data 
```

```python
# Extract the year from the 'Date' column using .loc to avoid the SettingWithCopyWarning
filtered_data.loc[:, 'Year'] = filtered_data['Date'].dt.year

# Group by year and count occurrences in FTHG and FTAG
bayern_summary = filtered_data.groupby('Year').agg(
    TotalGoalsAsHomeTeam=('FTHG', 'sum'),
    TotalGoalsAsAwayTeam=('FTAG', 'sum')
).reset_index()

# DataFrame
bayern_summary
```

```python
bayern_summary.dtypes
```

```python
bayern_summary.to_csv('bayern_summary.csv', index=False)
```

```python
import pandas as pd

data = {'column_to_split': ["STATIONS_ID;MESS_DATUM;QN_3;FX;FM;QN_4;RSK;RSKF;SDK;SHK_TAG;NM;VPM;PM;TMK;UPM;TXK;TNK;TGK;eor"]}
df = pd.DataFrame(data)

# Remove quotes and split the column into multiple columns
df_split = df['column_to_split'].str.strip("'").str.split(';', expand=True)

# Concatenate the original DataFrame with the new columns
df = pd.concat([df, df_split], axis=1)

# Drop the original column if needed
df = df.drop('column_to_split', axis=1)

# Display the resulting DataFrame
print(df)
```
