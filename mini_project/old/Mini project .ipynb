```python
import pandas as pd
bayern_data = pd.read_csv("Bundesliga_Results.csv")
bayern_data
```

----INSTRUCTIONS----

If column Home Team = Bayern Munich THEN count FTHG

If column Away Team = Bayern Munich THEN count FTAG

FTHG + FTAG = the number of goals Bayern Munich have scored per season in September





```python
# Filter rows where Bayern Munich is either the HomeTeam or AwayTeam
bayern_data = bayern_data[(bayern_data['HomeTeam'] == 'Bayern Munich') | (bayern_data['AwayTeam'] == 'Bayern Munich')]

# Count occurrences in FTHG and FTAG when Bayern Munich is the HomeTeam
bayern_home_goals = bayern_data.loc[bayern_data['HomeTeam'] == 'Bayern Munich', 'FTHG'].sum()

# Count occurrences in FTHG and FTAG when Bayern Munich is the AwayTeam
bayern_away_goals = bayern_data.loc[bayern_data['AwayTeam'] == 'Bayern Munich', 'FTAG'].sum()

# Create a new DataFrame with relevant data
bayern_summary = pd.DataFrame({
    'Team': ['Bayern Munich'],
    'TotalGoalsAsHomeTeam': [bayern_home_goals],
    'TotalGoalsAsAwayTeam': [bayern_away_goals]
})

# Display the summary DataFrame
bayern_summary
```

```python
import pandas as pd

# Extract the year from the 'Date' column
bayern_data['Year'] = pd.to_datetime(bayern_data['Date']).dt.year

# Filter rows where Bayern Munich is either the HomeTeam or AwayTeam
bayern_data = bayern_data[(bayern_data['HomeTeam'] == 'Bayern Munich') | (bayern_data['AwayTeam'] == 'Bayern Munich')]

# Group by year and count occurrences in FTHG and FTAG
bayern_summary = bayern_data.groupby('Year').agg(
    TotalGoalsAsHomeTeam=('FTHG', 'sum'),
    TotalGoalsAsAwayTeam=('FTAG', 'sum')
).reset_index()

# Display the summary DataFrame
bayern_summary
```
